```python
from math import radians, cos, sin, asin, sqrt

def haversine(coord1, coord2):
    # Convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(radians, [coord1[1], coord1[0], coord2[1], coord2[0]])
    
    # Haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371  # Radius of Earth in kilometers
    return c * r
# Example: Calculate distances from CCAFS LC-40 to key proximities
launch_site = [28.562302, -80.577356]  # CCAFS LC-40

proximities = {
    'coastline': [28.562302, -80.567356],     # Example coastline
    'railway':   [28.573255, -80.646895],
    'highway':   [28.563197, -80.576820],
    'city':      [28.3922, -80.6077]
}

# Calculate and print distances
for name, location in proximities.items():
    distance = haversine(launch_site, location)
    print(f"Distance from launch site to {name}: {distance:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
